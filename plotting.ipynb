```python
import matplotlib
df=pd.read_csv('scatterplot_data.csv')
df.dropna()
fig, ax = plt.subplots()
df['x']=(df.Nav+df.Manip+df.Percep+df.NL)*df.Real_sim
df['y'] = df.Num_scenes*df.Num_tasks
df['ylog']=np.log10(df['y'])
ax = sns.scatterplot(data=df, x='x', y='ylog', color='grey')

data = [[8,np.log10(30*574)]]
newdf = pd.DataFrame(data, columns=['x', 'ylog'])

sns.scatterplot(data=newdf, x='x', y='ylog', color='red', marker='*',size = 50, sizes = (500, 1000),)
plt.xlabel('Capabilities (Input Diversity)')
plt.ylabel('Complexity (Task + Scene Diversity)')

df['geommean']=round(np.sqrt(df['x']*df['ylog']),1)

for i, txt in enumerate(df.geommean):
    ax.annotate(txt, (df.x[i]+0.1, df.ylog[i]-0.1), fontsize=10)
    if txt==5.8:
      ax.annotate('LaNMP \n (ours)', (df.x[i]-0.25, df.ylog[i]+0.3))
legend = ax.legend('')
legend.remove()
```
